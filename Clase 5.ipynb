```python
df = pd.read_excel('datasets/Canada.xlsx',
                engine = 'openpyxl',
                sheet_name = 'Canada by Citizenship',
                skiprows = range(20),
                 skipfooter = 2
                )
df.head()
```

```python
df.columns
```

```python
df.drop(['Type','Coverage','AREA','REG','DEV','DevName','Unnamed: 43','Unnamed: 44','Unnamed: 45','Unnamed: 46','Unnamed: 47','Unnamed: 48','Unnamed: 49','Unnamed: 50'], 
        axis = 1,
       inplace = True)
df.columns
```

```python
df.rename(columns ={'OdName':'Country','AreaName':'Continent','RegName':'Region'},
         inplace = True)
```

```python
df.columns
```

## My variables of interest
'Country' - Represents the origin of the people. It is a categorical variable.

'Continent' - Represents the continent the people are from. It is a categorical variable.

'Region' - Represents the region inside the continent the people are from. It is a categorical variable.

'1980-2013' - Corresponds to the year the data belongs to. It is a continous variable.

```python
df.columns = list(map(str, df.columns))
df.columns
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
df['Total'] = df.iloc[:, 2:].sum(axis = 1)
df.head(2)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
mpl.style.use('ggplot')
japan_data = df.loc['Japan', '1980':'2013']
japan_data.plot()
plt.ylabel('Number of people')
plt.xlabel('Year')
```

```python
indiachina_data = df.loc[['India','China'], '1980':'2013']
nic = indiachina_data.transpose()
nic.plot()
plt.xlabel('Year')
plt.ylabel('Immigrants')
plt.title('Indian and Chinese Immigrants to Canada from 1980 to 2013')
```

## Conclusion:
We see a similar immigration pattern in both China and India
