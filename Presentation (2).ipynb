## Description of dataset: NBA Finals and MVPs

This dataset contains information about the NBA Finals participants, the score, the champion, and each year's MVP (Including their name, position, team, height in meters, height in feet, nationality and thei status at the end of the season) from 1950 to 2018.

## Description of variables

The variables of this dataset are:

    - Year: Shows the year the NBA season happened.
    
    - Western Champion: Shows the name of the team that won the Western Conference Championship. It is a categorical variable.
    
    - Eastern Champion: Shows the name of the team that won the Western Conference Championship.It is a categorical variable.
    
    - Result: Displays the result of the NBA finals from each year. It is a categorical variable.
    
    - NBA Champion: Shows the name of the team that won the NBA Finals. It is a categorical variable.
    
    - NBA Vice-Champion (Runner Up): Shows the name of the team that lost the NBA Finals each year. It is a categorical variable.
    
    - Sweep?: Displays whether the NBA Finals resulted in a 4-0. It is a Boolean variable.
    
    - MVP Name: Displays the initial and last name of the NBA MVP from each year. It is a categorical variable.
    
    - MVP Height in meters: Displays the MVP's height in meters. It is a numeric variable.
    
    - MVP Height in feet: Displays the MVP's height in feet. It is a numeric variable.
    
    - MVP Position: Displays the MVP's position in basketball. It is a categorical variable.
    
    - MVP Team: Displays the MVP's NBA team name. It is a categorical variable.
    
    - MVP Nationality: Displays the nationality of the MVP. It is a categorical variable.
    
    - MVP Status after the season: Displays whether the MVP from each year ended up as a champion, vice-champion or if he did not reach the NBA Finals. It is a categorical variable.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('NBA Finals and MVP.xlsx',
                engine = 'openpyxl')
df.head()
```

### First we display our data to have a first look at our dataset and see if we should get rid of any variables that are not necessary.

```python
pd.set_option('display.max_columns',500, 'display.max_rows', 500)
df
```

### We display the whole dataset to see if any other columns or rows show null values.

```python
df.drop([0,1,2,3,4,5], 
        axis = 0,
       inplace = True)
df.head()
```

#### We drop rows 0, 1, 2, 3, 4 and 5 because they have null values in all of the MVP-related variables.

```python
df.drop(['Final Sweep ?','MVP Height (ft)'],
       axis = 1,
       inplace = True)
df.head()
```

### Then we make the decision to drop columns 'Final Sweep ?'and 'MVP Height (ft)' because they are not necessary. We have the Result variable which will cause the Sweep variable to be true only if the result was a 4-0, and for the MVP Height (ft) variable we do not consider it necessary because we already have MVP Height in meters.

## What is the average height of each team's NBA MVP from 1956 to 2018?

```python
pd.pivot_table(df, 
               index ='MVP Team',
               values = 'MVP Height (m)',
               aggfunc = 'mean'
              )
```

### The table above shows the average height of each team's respective MVPs from 1956 to 2018. We can see the Philadelphia Warriors have the highest average with 2.16 meters.

## Which teams won the NBA Championship from 2010 to 2018?

```python
df.set_index('Year', inplace = True)
df.loc['2010':'2018','NBA Champion']
```

### The NBA Champions from 2010 to 2018 are Lakers in 2010, Mavericks in 2011, Heat in 2012 and 2013, Spurs in 2014, Cavaliers in 2016, and Warriors in 2015, 2017 and 2018. This shows us that the Golden State Warriors are the team that won the most championships from 2010 to 2018.

## Display an histogram of the heights of the MVPs from 1956 to 2018.

```python
hist_plot = df['MVP Height (m)'].plot(kind = 'hist')
plt.title('Heights of MVPs in meters')
```

### This histogram displays how the heights of the NBA MVPs is distributed. As we can see in the graph, most of the MVPs are above 2 meters tall.

## Display the tallest and shortest MVPs for each status.

```python
pd.pivot_table(df, 
               index = 'MVP status',
               values = 'MVP Height (m)',
               aggfunc = ['max','min']
              )
```

### This pivot table displays the maximum and minimum heights of each MVP status from 1956 to 2018. The tallest MVP to be a Champion, Vice-Champion and not reach a final was 2.18 meters tall. On the other side, the shortest MVP to win a championship was 1.85 meters tall; the shortest MVP to be a Vice-Champion was 1.83 meters tall; and the shortest MVP to not reach a final was 1.91 meters tall.
