```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head()
```

```python
sns.lmplot(x = 'Wingspan',
          y = 'CWDistance',
          data = df,
          fit_reg = False,
          hue = 'Gender')
plt.title('Scatter plot of CWDistance vs Wingspan', fontsize = 17)
plt.show()
```

```python
sns.swarmplot(x = 'Gender',
            y = 'CWDistance',
            data = df)
plt.show()
```

```python
sns.boxplot(data = df.loc[:,['Age', 'Height', 'Wingspan','CWDistance', 'Score']])
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Box plot representation of the entire data', fontsize = 15)
plt.show()
```

```python
sns.boxplot(data = df.loc[df['Gender']=='M',['Age', 'Height', 'Wingspan','CWDistance', 'Score']])
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Box plot representation of the MALE data', fontsize = 15)
plt.show()
```

1. Make a comparison of scores of Male and Female

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,4,1) # rows, columns, position
sns.boxplot(data = df.loc[df['Gender']=='M',['Score']])
                          
ax2 = fig.add_subplot(2,4,2)
sns.boxplot(data = df.loc[df['Gender']=='F',['Score']])

ax3 = fig.add_subplot(2,4,3)
sns.boxplot(data = df.loc[df['Gender']=='F',['Score']])

ax4 = fig.add_subplot(2,4,4)
sns.boxplot(data = df.loc[df['Gender']=='F',['Score']])
                          
ax1.title.set_text('Male box plot')
ax2.title.set_text('Female box plot')
ax3.title.set_text('Male box plot')
ax4.title.set_text('Female box plot')

plt.show()  
```
