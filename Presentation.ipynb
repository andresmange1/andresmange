## NBA Finals and MVPs

This dataset contains information about the NBA Finals participants, the score, the champion, and each year's MVP from 1950 to 2018

The variables of this dataset are:

    - Year
    
    - Western Champion
    
    - Eastern Champion
    
    - Result
    
    - NBA Champion
    
    - NBA Vice-Champion (Runner Up)
    
    - Sweep?
    
    - MVP Name
    
    - MVP Height in meters
    
    - MVP Height in feet
    
    - MVP Position
    
    - MVP Team
    
    - MVP Nationality
    
    - MVP Status after the season

```python
df = pd.read_excel('NBA Finals and MVP.xlsx',
                engine = 'openpyxl')
df.head()
```

```python
pd.set_option('display.max_columns',500, 'display.max_rows', 500)
df
```

```python
df.drop([0,1,2,3,4,5], 
        axis = 0,
       inplace = True)
df.head()
```

```python
df.drop(['Final Sweep ?','MVP Height (ft)'],
       axis = 1,
       inplace = True)
df.head()
```

## What is the average height of each team's NBA MVP from 1956 to 2018?

```python
pd.pivot_table(df, 
               index ='MVP Team',
               values = 'MVP Height (m)',
               aggfunc = 'mean'
              )
```

## Which teams won the NBA Championship from 2010 to 2018?

```python
df.set_index('Year', inplace = True)
df.loc['2010':'2018','NBA Champion']
```

## Display all the information about the 2001 NBA MVP.

```python
df.loc[2001, ['MVP Name','MVP Position','MVP Team','MVP Height (m)', 'MVP Nationality','MVP status']]
```

## Display an histogram of the heights of the MVPs from 1956 to 2018.

```python
hist_plot = df['MVP Height (m)'].plot(kind = 'hist')
```

```python
df['MVP Height (m)'].plot(kind = 'line')
plt.title('Height of MVPs', fontsize = 17)
```

## Display the tallest and shortest MVPs for each status.

```python
pd.pivot_table(df, 
               index = 'MVP status',
               values = 'MVP Height (m)',
               aggfunc = ['max','min']
              )
```
